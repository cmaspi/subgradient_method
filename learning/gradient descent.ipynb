```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use("dark_background")
```

```python
def abs(x):
    # return x**2
    return np.abs(x)

xx = np.linspace(-20,20,1000)
yy = abs(xx)

def mainFunction():
    fig = plt.figure()
    ax = fig.add_subplot()
    ax.plot(xx,yy, label='$y=x^2$')
    ax.legend()
    return fig,ax
```

```python
fig,ax= mainFunction()

# x = np.random.random(1)*15
x = np.array([6.6])
gamma = 10
nsteps=0
a = 0 
while np.abs(x[-1])>1e-2 and nsteps<100:
    nsteps += 1
    delta = np.sign(x[-1])
    a += np.abs(delta)
    x=np.append(x,x[-1]-delta/a*gamma)
    
print(np.abs(x[-1]))
print(nsteps)
ax.plot(x,abs(x), color='yellow',marker='o')
```
